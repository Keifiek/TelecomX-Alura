#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
datos_main = pd.read_json(url)
datos_main.info()
datos_main.head()
```

```python
datos_customer = pd.json_normalize(datos_main['customer'])
datos_customer.info()
datos_customer.head(10)
```

```python
datos_phone = pd.json_normalize(datos_main['phone'])
datos_phone.info()
datos_phone.head(10)
```

```python
datos_internet = pd.json_normalize(datos_main['internet'])
datos_internet.info()
datos_internet.head(10)
```

```python
datos_account = pd.json_normalize(datos_main['account'])
datos_account.info()
datos_account.head(10)
```

#ðŸ”§ TransformaciÃ³n

## **Transformacion de dataframe main**

```python
datos_main['Churn'] = datos_main['Churn'].map({'Yes' : True, 'No': False})
datos_main['Churn'] = datos_main['Churn'].astype(bool)
datos_main['customerID'] = datos_main['customerID'].astype('string')
datos_main.head(10)
datos_main.info()
```

## **Transformacion de dataframe customer**

```python
datos_customer[['Dependents', 'Partner']] = datos_customer[['Dependents', 'Partner']].map(lambda x: True if x == 'Yes' else False)
datos_customer[['Dependents', 'Partner']] = datos_customer[['Dependents', 'Partner']].astype(bool)
datos_customer.info()
datos_customer.head(10)
```

## **Transformacion de dataframe phone**

```python
datos_phone[['PhoneService', 'MultipleLines']] = datos_phone[['PhoneService', 'MultipleLines']].map(lambda x: True if x == 'Yes' else False)
datos_phone[['PhoneService', 'MultipleLines']] = datos_phone[['PhoneService', 'MultipleLines']].astype(bool)
datos_phone.info()
datos_phone.head(10)
```

## **Transformacion de dataframe internet**

```python
datos_internet[['OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']] = datos_internet[['OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']].map(lambda x: True if x == 'Yes' else False)
datos_internet[['OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']] = datos_internet[['OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']].astype(bool)
datos_internet.info()
datos_internet.head(10)
```

##**Transformacion de dataframe account**

```python
datos_account['PaperlessBilling'] = datos_account['PaperlessBilling'].map({'Yes' : True, 'No': False})
datos_account['PaperlessBilling'] = datos_account['PaperlessBilling'].astype(bool)
datos_account['Charges.Total'] = datos_account['Charges.Total'].replace('[\$,]', '', regex=True)
datos_account['Charges.Total'] = pd.to_numeric(datos_account['Charges.Total'], errors='coerce')
datos_account['Charges.Total'] = datos_account['Charges.Total'].astype(np.float64)
datos_account.info()
datos_account.head(1000)
```

## **Crear dataframe unificado post-tranformacion de datos**

```python
columas_main = datos_main[['customerID', 'Churn']]
dataframe_unificado = pd.concat([columas_main,datos_customer,datos_phone,datos_internet,datos_account],axis=1)
dataframe_unificado
```

## **Crear columna cuentas diarias**

```python
datos_main['Cuentas_diarias'] = datos_account['Charges.Monthly']/30
dataframe_unificado['Cuentas_diarias'] = dataframe_unificado['Charges.Monthly']/30
```

## **Introduccion**


En este analisis se busca dar con la razon del porque la empresa Telecom X se encuentra actualmente sufriendo una alta evasion de clientes tambien conocida como Churn.
Para lograr este objetivo la empresa a solicitado una analisis de datos, proporcionando datos de los clientes asi como diferentes caracteristicas para contrastar entre si, como puede ser el genero, el tiempo contratado o el pago total generado, entre otros.


## **Limpieza y Tratamiento de Datos**

### **Importacion de datos**

Los datos fueron proporcionados por medio de un archivo JSON por lo cual se decidio usar la ayuda de la libreria pandas para representar los datos en un dataframe previo a la limpieza y tratamiento de los datos.
Lo primero que se observo al intentar generar el dataframe con ayuda de la libreria pandas es que teniamos varios niveles dentro de nuestro archivo JSON por lo cual debiamos normalizar estos niveles en diferentes dataframes.  

#### **Normalizacion**  
Despues de notar que columnas de nuestro dataframe deberiamos de normalizar procedimoa a generar estos nuevos dataframes obteniendo 5 dataframes diferentes con el principal incluido.

### **Limpieza y tranformacion de datos**


#### **Limpieza**
En este caso no fue necesario hacer muchos pasos en la limpieza de los datos puesto que parecian seguir en su mayoria formatos establecidos, sin embargo en la columa de valor total me di cuenta que habia algunos valores con espacios que posteriormente daria problemas al intentar convertir a un tipo float por lo cual que hico una tratamiento preliminar a la transformacion

#### **Transformacion de Datos**
Para este paso si hubo bastante carga de cosas por realizar puesto que en casi todos los dataframes generados los tipos de datos de las columnas eran en su mayoria de tipo object y no tenian los tipos bien definidos, para arreglar esto primero converti algunos datos que note que eran a grandes rasgos booleanos.

##### **Booleanos**
En el dataframe encontramos varios tipos de datos que a simple vista notamos que pueden ser considerados booleanos puesto que solo tienen dos tipos de valores "Yes" y "No" para esto hicimos un map a todos estos valores para cambiarlos a True o False y posteriormente hicimos que pandas lo detectara como un boolano con ayuda de la funcion astype.

##### **Flotantes**
En este caso el unico que no detecto como flotante fue la columna de gasto total por lo cual hice algunos cambios como quitar las comas, los espacios y hacerlo de manera regresiva con un regex en caso de que se hubiera ido un espacio al inicio y otro al final, los datos quedaron estandarizados como los demas flotantes.

##### **String**

En este caso no hubo mucho que hacer puesto que a pesar de que aplicara el astype con str el pandas lo seguia reconociendo como un object por lo cual decidi usar string solo para recordarlo pero no fue del todo necesario posteriormente.


##### **Dataframes**
Como se menciono durante la importacion se ocuparon crear varios dataframes para poder normalizar algunas columnas sin embargo me di cuenta que esto generaria problematicas para el analisis posterior por lo cual revise y todos los dataframes geenrados tenian una correspondencia de 1 a 1, es decir que teniamos la misma cantidad de registros, por lo cual decidi concatenar todos los dataframes generados obteniendo un nuevo dataframe unificado que era lo que me interesaba conseguir al inico.
